```python
datagen = keras.preprocessing.image.ImageDataGenerator(rescale=1.0/255.0)
```

```python
mybatch
```

```python
train_generator = datagen.flow_from_dataframe(dataframe=z1,
                                       directory=None,
                                       x_col = 'id',
                                       y_col='label',
                                       class_mode = 'categorical',
                                       batch_size = mybatch,
                                       shuffle=True,
                                       target_size= (mysize,mysize)
                                      )
```

```python
val_datagen = keras.preprocessing.image.ImageDataGenerator(rescale=1.0/255.0)
```

```python
val_generator = val_datagen.flow_from_dataframe(dataframe=z2,
                                       directory=None,
                                       x_col = 'id',
                                       y_col='label',
                                       class_mode = 'categorical',
                                       batch_size = mybatch,
                                       target_size= (mysize,mysize)
                                      )
```

```python
test_datagen = keras.preprocessing.image.ImageDataGenerator(rescale=1.0/255.0)
```

```python
test_generator = test_datagen.flow_from_dataframe(dataframe=testfiles,
                                       directory=None,
                                       x_col = 'id',
                                       #y_col='label',
                                       class_mode = None,
                                       batch_size = 1,
                                       shuffle=False,
                                       target_size= (mysize,mysize)
                                      )
```

```python
start=datetime.datetime.now()
model = keras.Sequential([
    keras.layers.Conv2D(32, kernel_size=3, activation=  'relu', input_shape=(mysize,mysize,3)),
    keras.layers.Conv2D(32, kernel_size=3, activation = 'relu'),
    keras.layers.MaxPooling2D(pool_size=(2,2)),
    keras.layers.Dropout(0.25),
    
    keras.layers.Conv2D(64, kernel_size=3, activation = 'relu'),
    keras.layers.Conv2D(64, kernel_size=3, activation = 'relu'),
    keras.layers.MaxPooling2D(pool_size=(2,2)),
    keras.layers.Dropout(0.25),
    
    keras.layers.Conv2D(128, kernel_size=3, activation = 'relu'),
    keras.layers.Conv2D(128, kernel_size=3, activation = 'relu'),
    keras.layers.MaxPooling2D(pool_size=(2,2)),
    keras.layers.Dropout(0.25),
    
    keras.layers.Flatten(),
    keras.layers.Dense(256,activation ='relu'),
    keras.layers.Dropout(0.5),
    
    keras.layers.Dense(2,activation='softmax')
])
end=datetime.datetime.now()
print(end-start)
```

```python
model.compile(optimizer=keras.optimizers.Adam(lr=.0001, beta_1=.5, beta_2=0.999),
              loss=keras.losses.binary_crossentropy,
              metrics=['binary_accuracy'])
```

```python
model.summary()
```

```python
mysave = keras.callbacks.ModelCheckpoint('mymod.hdf5', save_best_only = True,monitor='val_loss',mode='min')
```

```python
start=datetime.datetime.now()
my_train = model.fit_generator(train_generator,
                     epochs =15,
                     callbacks=[mysave],
                     validation_data = val_generator,
                     validation_steps = valsteps,
                     steps_per_epoch=trainingsteps
                    )
model.save('./cata_and_dogs/my_model.h5')
end=datetime.datetime.now()
```
